Import Libraries needed:

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier 
import matplotlib.pyplot as plt
```

Load the dataset into a variable

```python
breast_cancer_data = load_breast_cancer()
```

Split the dataset into training and validation sets

```python
training_data, validation_data, training_labels, validation_labels = train_test_split(breast_cancer_data.data, breast_cancer_data.target, test_size = 0.2, random_state = 100)
```

Created an empty list to store classifier scores

```python
accuracies = []
```

Run the classifier for k from 1 to 100 and append score to accuracies

```python
for i in range(1,101):
  classifier = KNeighborsClassifier(n_neighbors = i)
  classifier.fit(training_data, training_labels)
  #print(classifier.score(validation_data, validation_labels))
  accuracies.append(classifier.score(validation_data, validation_labels))
```

Create a list from 1 to 100

```python
k_list = list(range(1, 101))
```

Plot the values of k to see what values of k provide the most accuracy

```python
plt.plot(k_list, accuracies)
plt.xlabel('K Value')
plt.ylabel('Validation Accuracy')
plt.title('Breast Cancer Classifier Accuracy')
plt.show()
```
